<h1>Clustering</h1>

\begin{equation*}
\underset{S}{\operatorname{argmin}} \sum_{i-1}^{k}\sum_{x \in Si}\| x - \mu_{i} \|^2 = \underset{S}{\operatorname{argmin}}
\sum_{i-1}^{k}|S_i|VarS_i
\end{equation*}

\begin{equation*}
\underset{S}{\operatorname{argmin}} \sum_{i-1}^{k}\sum_{x \in Si}\| x - \mu_{i} \|^2 = \underset{S}{\operatorname{argmin}}
\sum_{i-1}^{k}|S_i|VarS_i
\end{equation*}

Given a set of observations $(x_1, x_2, …, x_n)$, where each observation is a d-dimensional real vector, k-means clustering aims to partition the n observations into k (≤ n) sets S = {S1, S2, …, Sk} so as to minimize the within-cluster sum of squares (WCSS) (i.e. variance). Formally, the objective is to find:

\begin{equation*}
{\displaystyle {\underset {\mathbf {S} }{\operatorname {arg\,min} }}\sum _{i=1}^{k}\sum _{\mathbf {x} \in S_{i}}\left\|\mathbf {x} -{\boldsymbol {\mu }}_{i}\right\|^{2}={\underset {\mathbf {S} }{\operatorname {arg\,min} }}\sum _{i=1}^{k}|S_{i}|\operatorname {Var} S_{i}}
\end{equation*}

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array([[1,2], [1,4], [1,0], [10,2], [10,4], [10,0]])
```

```python
print(X)
```

```python
x1 = [item[0] for item in X]
x1
```

```python
x2 = [item[1] for item in X]
x2
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
kmeans.predict([[0, 0], [12, 3]])
```

```python
kmeans.cluster_centers_
```

```python
c1 = [10., 2.]
c2 = [1., 2.]
plt.plot(x1, x2, 'ro', c1[0], c1[1], 'bs', c2[0], c2[1], 'g^')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
th = np.array([[83,86], [64,65], [72,90], [81,75], [70,96], [68,80], [65,70], 
              [75,80], [71,91], [85,85], [80,90], [72,95], [69,70], [75,70]])
#th
```

```python
t = [item[0] for item in th]
h = [item[1] for item in th]
#print(t,h)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(th)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
c = kmeans.cluster_centers_

plt.plot(t, h, 'ro', c[0][0], c[0][1], 'bs', c[1][0], c[1][1], 'g^')
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(th)
```

```python
kmeans.cluster_centers_
```

```python
i_s = 0
i_f = len(t)
l = kmeans.labels_
s1 = ['ro', 'bo', 'go']
s2 = ['gs', 'rs', 'b^']
c = kmeans.cluster_centers_

for i in range(i_s, i_f):
    plt.plot(t[i], h[i], s1[l[i]])
    
for i in range(0, 3):
    plt.plot(c[i][0], c[i][1], s2[l[i]])
    
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.show()
```
